## seaborn and this will be the end for data visualization

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## importing built in data set of seaborn

```python
print(sns.get_dataset_names()) 
crash_df = sns.load_dataset('car_crashes')
print(crash_df.head())
```

```python
var = [1,2,3,4,5,6]
var_1 = [2,3,4,10,6,7]
plt.plot(var,var_1)
plt.show()
```

```python
var = [1,2,3,4,5,6]
var_1 = [2,3,4,10,6,7]
x = pd.DataFrame({"var": var, "var_1":var_1})
sns.lineplot(x = 'var', y = 'var_1', data=x)
# the differnce as we can see is here scales are labeled
```

```python
y_1 = sns.load_dataset('penguins').head(20)
print(y_1.head())
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = y_1, hue='sex', size=20,style = 'sex',palette= 'pink',
              markers="o",legend=False)
```

## histogram plot

```python
var = sns.load_dataset("penguins")
var.head(10)
```

```python
sns.displot(var["bill_length_mm"],bins = [30,40,50,60],kde = False, rug = True , color="hotpink") #log_scale=True) 
plt.show()
# rug is for the below scale
# bins are for the range
# kdde is the line witch follows the head like tradding
```

## barplot

```python
var = sns.load_dataset("penguins")
var.head(10)
```

```python
sns.barplot(x = var.island,y = var.bill_depth_mm, hue = var.sex, order=['Dream','Torgersen','Biscoe'],
            hue_order=['Female','Male'])
# we can change the order of the plot by using oder parameter
```

```python
# horizontal
sns.barplot(x = var.bill_length_mm,y = var.bill_depth_mm,orient='h')
```

## scatter plot

```python
sns.scatterplot(x = "bill_length_mm", y = 'bill_depth_mm',hue= "sex", data=var)
```
