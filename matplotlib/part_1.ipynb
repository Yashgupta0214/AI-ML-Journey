## matplotlip

```python
import matplotlib.pyplot as plt
import numpy as np
```

pyplot is a sub module for matplotlib

## we are drawing two lines wiz a=x- ais sand y- axis

```python
xpoints = np.array([0,10])
ypoints = np.array([0,10])
plt.plot(xpoints,ypoints)
plt.show
```

no line in between the graph

```python
xpoints = np.array([0,10])
ypoints = np.array([0,10])
plt.plot(xpoints,ypoints,"o")
plt.show
```

zig zag pattern in graph

```python
xpoints = np.array([0,2,3,5,10])
ypoints = np.array([0,4,2,1,10])
plt.plot(xpoints,ypoints)
plt.show

# if we do not specify x axis it automatically makes it 1,2,3,.....
```

marker 

```python
ypoints = np.array([2,5,2,10])
plt.plot(ypoints, marker = "o")
plt.show
```

markoing colour to the line

```python
ypoints = np.array([2,8,3,10])
plt.plot(ypoints,"o-.r", marker = "o", ms = 10, mec = "r", mfc = "b" )# we can choose many colours
plt.show

# line reference
# : - dotted line
#- solid line
# -- dashed line
# -. dashed,dotted line
# 
# marker size : it is the big red circle
#  mec : blue out line
# mfc : marker face colour
```

## line style

```python
# used to change the sstyle of the plotted line
ypoints = np.array([3,8,4,10])
plt.plot(ypoints,linestyle = "dashed", color = "red", linewidth = 5)#  we can also use ls instead of linestyle
plt.show()
# line colour use colour in the plot syntax
# line width use linewidth in plot syntax
```

plot multiple line

```python
xpoints = np.array([0,2,3,5,10])
ypoints = np.array([0,4,2,1,10])
plt.plot(xpoints)
plt.plot(ypoints)
plt.show
```

## titles and labels on a graph

```python
xpoints = np.array([0,2,3,5,10])
ypoints = np.array([0,4,2,1,10])
plt.title("title in a graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xpoints,ypoints)
plt.show()
```

# fonts for title and label

```python
xpoints = np.array([0,2,3,5,10])
ypoints = np.array([0,4,2,1,10])
#font1 = {"family : serif","color": 'blue' ,"size": 20}
plt.title("title in a graph")
plt.xlabel("x axis") #fontdict=font1
plt.ylabel("y axis")# we can also set the location of the label by loc = left / right / 
plt.plot(xpoints,ypoints)
plt.show()
```

## GRID LINES

```python
xpoints = np.array([0,2,3,5,10])
ypoints = np.array([0,4,2,1,10])
plt.title("title in a graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xpoints,ypoints)
plt.grid(axis="x",color="red", linestyle = "dashed", linewidth = 5) # to show grid lines 
plt.show()

# display grid lines only to x- axis or y axis
# color of grid lines "use color in grid
# we can also chhose the line style by "using line style" and decide the line width
```
