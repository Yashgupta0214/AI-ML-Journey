# bar charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

## horizontal bars

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.barh(x,y, color = "red") # just have to write barh instead of bar
plt.show()
#colors c =red
```

## width and height of a bar

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y, width=0.3)
plt.show()
# we use height for horizontal bars
```

## histogram

```python
x = np.random.normal(170,10,250)
plt.hist(x, color="red")
plt.show()
```

## pie diagram

```python
X = np.array([32,25,15,5])
my_label =  ['mango','banana','ganna','grapes']

plt.pie(X, labels=my_label,startangle=180)
plt.show()
# labeling the diff colors and value
# we can also choose the start angle here it starts from zero degree
```

## exploding a pie chart

```python
X = np.array([32,25,15,5])
my_label =  ['mango','banana','ganna','grapes']
my_explode = [0,0.3,0,0.3]
plt.pie(X, labels=my_label,startangle=180, explode = my_explode, shadow=True)
plt.show()
# we can also give the shadow by using shadow parameter
```

## each colour for each wedge

```python
X = np.array([32,25,15,5])
my_label =  ['mango','banana','ganna','grapes']
my_colors = ['orange','red','hotpink','blue']
plt.pie(X, labels=my_label,startangle=180,colors=my_colors)
plt.legend(title = "fruits")# it explanes in a box
plt.show()
```
