## subplot , didsplay multiple plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.Subplot(2,1,1) # the figure has 1 for rows 2 for colum and 1 for this plot respectivily(1,2,1)
plt.plot(x,y)


#plot2
x = np.array([1,1,2,3])
y = np.array([10,20,30,40])
plt.Subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

## scatter function

```python
x = np.array([1,23,43,45,55])
y = np.array([27,11,42,41,21])
plt.scatter(x,y)
plt.show
```

##  comparing 2 scatter chart

```python
x = np.array([1,23,43,45,55])
y = np.array([27,11,42,41,21])
plt.scatterbg(x,y, color = "hotpink")

x = np.array([12,14,19,27,39])
y = np.array([49,22,18,7,32])
plt.scatter(x,y, color = "green")

plt.show()
```

SIZE OF SCATTER PLOT

```python
x = np.array([1,23,43,45,55])
y = np.array([27,11,42,41,21])
size =  30
plt.scatter(x,y, color = "hotpink", s= size)
size =  60
x = np.array([12,14,19,27,39])
y = np.array([49,22,18,7,32])
plt.scatter(x,y, color = "green", s=size)
plt.show()
```

```python
# transpaerncy of the dots by using alpha
size =  60
x = np.array([12,14,19,27,39])
y = np.array([49,22,18,7,32])
plt.scatter(x,y, color = "green", s=size, alpha=0.1)
plt.show()
```

## combinig colour size alpha 

```python
x = np.random.randint(100,size = (100))
y = np.random.randint(100,size = (100))
colors = np.random.randint(100,size = (100))
sizes = 10*np.random.randint(100,size = (100))
plt.scatter(x,y, c = colors, s = sizes, alpha=0.4, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
