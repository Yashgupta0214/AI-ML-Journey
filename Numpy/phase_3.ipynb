## advance operation with bussiness example

```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== zomato sales analysis ===")
print("shape of structure: ",sales_data.shape)
print("first three restaurent: ", sales_data[:3])
```

```python
#total sales per year

print(np.sum(sales_data[:,1:], axis=0))

# minimum sales per year

min_sales = np.min(sales_data[:,1:], axis=1)
print(" min sales per year",min_sales)

# o axis means y axis and 1 means x axis

#maximum sales per year

max_sales= np.max(sales_data[:,1:], axis=0)
print( "max sales per year",max_sales)

#avg sales per year

avg_sales = np.mean(sales_data[:,1:], axis = 1)
print("average sale per restaurent", avg_sales)
```

```python
# cummuulative sales

cum_sum = np.cumsum(sales_data[:,1:], axis=1)
print(cum_sum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cum_sum , axis=0))
plt.title("average cummulative frequency")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show

```

```python
vector_1 = np.array([1,2,3,4,5])
vector_2 = np.array([6,7,8,9,10])

print("vector addition", vector_1+vector_2)

print("vector multiplication", vector_1*vector_2)

print("dot product", np.dot(vector_1,vector_2))

angle = np.arccos(np.dot(vector_1,vector_2)/(np.linalg.norm(vector_1)* np.linalg.norm(vector_2)))
print(angle)
```
